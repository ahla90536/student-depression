```python
!pip install scikit-learn
from sklearn.preprocessing import LabelEncoder
```

```python
encoder=LabelEncoder()
```

```python
df['Gender']=encoder.fit_transform(df['Gender'])
```

```python
df['Have you ever had suicidal thoughts ?']=encoder.fit_transform(df['Have you ever had suicidal thoughts ?'])
df['Family History of Mental Illness']=encoder.fit_transform(df['Family History of Mental Illness'])
df['Sleep Duration']=encoder.fit_transform(df['Sleep Duration'])
df['Dietary Habits']=encoder.fit_transform(df['Dietary Habits'])
```

```python
df_ = pd.get_dummies(df, columns=['City', 'Profession', 'Degree'])
```

```python
from sklearn.preprocessing import OneHotEncoder
ohe=OneHotEncoder()
```

```python
df_
```

```python
x=df_.drop(columns=['Depression'])
```

```python
y=df_['Depression']
```

```python
from sklearn.model_selection import train_test_split
```

```python
xtrain,xtest,ytrain,ytest=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler=StandardScaler()
```

```python
xtrain_scaled=scaler.fit_transform(xtrain)
```

```python
from sklearn.tree import DecisionTreeClassifier

model = DecisionTreeClassifier(random_state=42,max_depth=4)
model.fit(xtrain_scaled, ytrain)
```

```python
xtest_scaled=scaler.fit_transform(xtest)
```

```python
ypred=model.predict(xtest_scaled)
```

```python
ypred
```

```python
ytest
```

```python
from sklearn.metrics import accuracy_score
```

```python
accuracy_score(ytest,ypred)
```

```python
sample = xtest.iloc[[100]]

prediction = model.predict(sample)

print("Predicted:", prediction[0])
print("Actual:", ytest.iloc[100])
```
